```python
# 리스트이름 = [[0] * 열 for i in range(행)] => [0]이 열의 개수만큼 반복되는 리스트를 행의 개수만큼 만든다.
# data = [[0] * 5 for i in range(4)] # 모든 요소가 0으로 초기화된 4행 5열인 2차원 리스트가 만들어진다.
# print(data)
# print(len(data))    # 행의 개수
# print(len(data[0])) # 열의 개수
```

```python
i, j = map(int, input("2차원 리스트를 만들려하는 행과 열의 개수를 입력하세요 : ").split())
data = [[0] * j for i in range(i)]
# print(data)

n = 0 # 리스트에 채워질 숫자가 기억되는 변수를 선언한다.
# 전체 행의 개수 만큼 반복시킨다.
for i in range(len(data)):
    # 짝수행은 숫자가 왼쪽에서 오른쪽으로 채워지고 홀수행은 숫자가 오른쪽에서 왼쪽으로 채워진다.
    # 짝수행인가 홀수행인가 판단한다.
    if i % 2 == 0:
        # 짝수행은 왼쪽에서 오른쪽으로 숫자를 채운다.
        for j in range(len(data[0])): # 0, 1, 2, 3, 4
            n += 1
            data[i][j] = n
    else:
        # 홀수행은 오른쪽에서 왼쪽으로 숫자를 채운다.
        for j in range(len(data[0]) - 1, -1, -1): # 4, 3, 2, 1, 0
            n += 1
            data[i][j] = n

# 2차원 리스트 출력하기
for i in range(len(data)):
    for j in range(len(data[0])):
        print("%3d " % data[i][j], end = "")
    # 한 행을 출력했으므로 줄을 바꿔준다.
    print()
```

```python
i, j = map(int, input("2차원 리스트를 만들려하는 행과 열의 개수를 입력하세요 : ").split())
data = [[0] * j for i in range(i)]

n = 0                  # 리스트에 채워질 숫자가 기억되는 변수를 선언한다.
start = 0              # 2차원 리스트에 숫자가 채워지기 시작하는 열
end = len(data[0]) - 1 # 2차원 리스트에 숫자가 채워지는 마지막 열
sw = 1                 # start 부터 end 까지 증가치로 사용할 변수

# i가 짝수일 경우 j는 0, 1, 2, 3, 4
# i가 홀수일 경우 j는 4, 3, 2, 1, 0

for i in range(len(data)):
    for j in range(start, end + sw, sw):
        n += 1
        data[i][j] = n
    # ===== for j in range(start, end + sw, sw): 끝 =====
    # 행(i)이 짝수일 때 start 부터 end + sw 까지 1씩 증가해야 지만 홀수가 되면 start 부터 end + sw 까지 1씩 감소해야 한다.
    # 리스트의 한 행에 숫자를 채운 후 start와 end에 저장된 값을 교환하고 증가치(sw)의 부호를 변경한다.
    start, end = end, start
    sw *= -1

for i in range(len(data)):
    for j in range(len(data[0])):
        print("%3d " % data[i][j], end = "")
    print()
```
